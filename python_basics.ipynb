```python
trip_data = [
    {"country": "Norway", "trip_days": 7, "recovery_score": 8},
    {"country": "Italy", "trip_days": 10, "recovery_score": 9},
    {"country": "France", "trip_days": 4, "recovery_score": 6},
    {"country": "Austria", "trip_days": 3, "recovery_score": 5}
]

total = sum(item["recovery_score"] for item in trip_data)
avg_recovery = total / len(trip_data)
avg_recovery
```

```python
days = [item["trip_days"] for item in trip_data]
recovery = [item["recovery_score"] for item in trip_data]

list(zip(days, recovery))
```
